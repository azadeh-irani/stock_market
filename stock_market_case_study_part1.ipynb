```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path=r'C:\Users\azade\OneDrive\Desktop\udemy_project_course\timeseries\individual_stocks_5yr'
company_list=['AAPL_data.csv','GOOG_data.csv','MSFT_data.csv','AMZN_data.csv']
all_data=pd.DataFrame()
for file in company_list:
    current_df=pd.read_csv(path+'/'+file)
    all_data=pd.concat([all_data,current_df])
all_data.shape
```

```python
all_data.head(5)
```

```python
####analyse closing price of all stocks####
```

```python
tech_list=all_data['Name'].unique()  ## retrieve the unique values in the 'Name' 
```

```python
all_data.dtypes #we have to change data type for date
```

```python
all_data['date']=pd.to_datetime(all_data['date']) #change date type
```

```python
plt.figure(figsize=(20,12))
for i,company in enumerate(tech_list,1): # a loop that iterates over the elements of the tech_list
    plt.subplot(2,2,i)
    df=all_data[all_data['Name']==company]  ##this code filters the all_data DataFrame to include only the rows where the 'Name' column matches the value stored in the company variable.
    plt.plot(df['date'],df['close'])
    plt.xticks(rotation='vertical')
    plt.title(company)



```
